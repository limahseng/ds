# Stack

```python
# Stack class
# using first principles
# array

class Stack(object):

    MAX = 10

    def __init__(self):
        self.stack = []
        for i in range(self.MAX):
            self.stack.append(0)
        self.top = -1

    def is_empty(self):
        return self.top == -1

    def is_full(self):
        return self.top == Stack.MAX-1

    def push(self, data):
        if self.is_full():
            print("Cannot insert to full stack")
            return -1
        else:
            self.top = self.top + 1
            self.stack[self.top] = data

    def pop(self):
        if self.is_empty():
            print("Cannot delete from empty stack")
            return -1
        else:
            data = self.stack[self.top]
            self.top = self.top - 1
            return data

    def peek(self):
        return self.stack[self.top]

    def display(self):
        for item in self.stack:
            print(item)
```

```python
# Stack class
# using first principles
# linked list

class Node(object):

    def __init__(self, data):
        self.data = data
        self.next = None


class Stack(object):

    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top == None

    def is_full(self):
        """ Usually not applicable unless system memory exhausted """
        pass

    def push(self, data):
        node = Node(data)
        node.next = self.top
        self.top = node

    def pop(self):
        if self.is_empty():
            print("Cannot delete from empty stack")
            return -1
        else:
            data = self.top.data
            self.top = self.top.next
            return data

    def peek(self):
        return self.top.data

    def display(self):
        curr = self.top
        while curr != None:
            print(curr.data)
            curr = curr.next
```
